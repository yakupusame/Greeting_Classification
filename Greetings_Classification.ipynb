```
sentences = df['Selected'].tolist()
len(sentences)
```

```
one_string_sentences = "".join(sentences)
one_string_sentences
```

```
import string
string.punctuation
```

```
df_punctuation_removed = [char for char in one_string_sentences if char not in string.punctuation]
df_punctuation_removed
```

```
df_removed = "".join(df_punctuation_removed)
df_removed
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```
df_removed = [word for word in df_removed.split() if word.lower() not in stopwords.words('english')]
df_removed
```

```
# Let's define a pipeline to clean up all the messages 
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords
def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```
df.head()
```

```
from sklearn.feature_extraction.text import CountVectorizer
# Define the cleaning pipeline we defined earlier
vectorizer = CountVectorizer(analyzer = message_cleaning)
df_all = vectorizer.fit_transform(df['Selected'])
```

```
print(df_all.toarray()) 
```

```
df_all.shape
```

```
df.head()
```

```
df.drop(['Selected'],axis = 1 ,inplace=True)
df_1 = pd.DataFrame(df_all.toarray())
df_final = pd.concat([df_1,df],axis = 1)
df_final
```

```
df_final.dropna(inplace=True)
```

```
x = df_final.drop(['Greeting'],axis=1)
x
```

```
y = df_final['Greeting']
y
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(x_train, y_train)
```

```
from sklearn.metrics import classification_report, confusion_matrix
y_pred = NB_classifier.predict(x_train)
y_pred
cm = confusion_matrix(y_train, y_pred)
cm
```
